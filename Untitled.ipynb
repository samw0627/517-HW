```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time range
t = np.linspace(-10, 10, 1000)

# Define the triangle waveform function f(t)
def f(t):
    return np.where((t >= -1) & (t <= 1), 1 - np.abs(t), 0)

# Define the square waveform function v(t)
def v(t):
    return np.where((t >= -5) & (t <= 0), 4, 0)

# Calculate the waveforms
ft = f(t)
vt = v(t)

# Calculate the convolution g(t)
gt = np.convolve(ft, vt, mode='same') / len(t)  # Normalize by the length of t

# Plot the waveforms
plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.plot(t, ft, label='f(t) (Triangle)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Triangle Waveform: f(t)')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, vt, label='v(t) (Square)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Square Waveform: v(t)')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, gt, label='g(t) (Convolution)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Convolution Waveform: g(t)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
